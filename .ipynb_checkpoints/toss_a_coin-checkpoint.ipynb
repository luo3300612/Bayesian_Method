## 贝叶斯方法
参考书《BAYESIAN METHODS FOR Hackers》by Cameron Davidson-Pilon

建立一个对bayes的直觉，即：

**随着证据更新信念**

频率学派基于大数定律，用频率来逼近概率，而贝叶斯统计学派以一个主观认知（初始信念）为起点，随着事件的发生（新的证据），来更新自己的信念。

初始信念即为先验概率，更新的信念即为后验概率，更新的信念继而能作为下一次事件（新的证据）发生前的先验概率，待下一次事件发生之后通过更新得到新的后验概率。

二者不同的区别在于：
* 频率学派从大量的重复实验来得到概率的估计，而贝叶斯学派在实验的不断发生中更新信念
* 频率学派得到的概率是一个确定的值，而贝叶斯学派会得到概率的一个分布

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
%matplotlib qt5
```

```python
coin = [0, 1]
density = 100
x = 1 / density * np.arange(density)
y = np.ones((density, )) * 1 / density

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'bo', animated=True)


def init():
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 0.12)
    return ln,


def update(frame):
    n = random.choice(coin)
    p_up = 0
    p_down = 0
    for i in range(y.shape[0]):
        p_up += x[i] * y[i]
        p_down += (1 - x[i]) * y[i]
    if n is 0:
        print('down')
        for i in range(y.shape[0]):
            y[i] = ((1 - x[i]) * y[i]) / p_down
    elif n is 1:
        print('up')
        for i in range(y.shape[0]):
            y[i] = x[i] * y[i] / p_up

    ln.set_data(x, y)
    return ln,


ani = FuncAnimation(
    fig,
    update,
    frames=np.linspace(0, 2 * np.pi, 128),
    init_func=init,
    blit=True)

plt.show()
```
